# Useful trick to change values in one column based on occurrence of a string in another (in python)

```python
# In python 3.7
import pandas as pd
from urllib.request import urlretrieve

# Set the url address (note the raw url from GitHub)
url = 'https://raw.githubusercontent.com/byounginger/Practice_code/master/Datasets/sample.csv'

# Save the url locally
urlretrieve(url, 'sample.csv')

# Read into pandas
df = pd.read_csv('sample.csv')

df

# Create boolean of column of interest that contains desired string
rows = df.Desc.str.contains('BUDGET=')

# Replace values in another column (Value) with those contained in the first column (Desc)
df.loc[rows,'Value'] = df.loc[rows, 'Desc']

# Check results
df
```
